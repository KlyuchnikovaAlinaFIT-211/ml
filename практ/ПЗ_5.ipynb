4. Объедините датафреймы из п. 2, 3 в один датафрейм, удалите дубликаты строк.

7. Поменяйте местами нулевую по порядку строку с элементом с индексом 2 (loc, iloc).

8. В столбце Sex замените female/male на f/m (map).

```python
def fm(x):
    if x=="female":
        return "f"
    elif x=="male":
        return "m"
new_dfm=new_df.copy(deep=True)   
new_dfm["sex"]=new_dfm["sex"].map(fm)
new_dfm
```

9. Найдите билеты по которым плыло 6 и более человек, отобразите список людей (groupby, count).

```python
dft = new_df.groupby(['ticket'])['ticket'].count().reset_index(name='counts')
dft1 = dft.loc[dft['counts'] > 5]
dft1 = dft1.reset_index(drop=True)
dft2 = new_df.loc[new_df['ticket'] == dft1.loc[0]['ticket']]
for i in range(1, len(dft1)):
    dft2 = pd.concat([dft2, new_df.loc[new_df['ticket'] == dft1.loc[i]['ticket']]])
dft2
```

10. Найдите людей, которые плыли в одной каюте с людьми из датафрейма из п1.


```python
dff = new_df.loc[new_df['name'] == df.loc[0]['Name']]
for i in range(1, len(df)):
    dff = pd.concat([dff, new_df.loc[new_df['name'] == df.loc[i]['Name']]])
dff = dff.reset_index(drop=True)
dff2 = new_df.loc[(new_df['cabin'] == dff.loc[0]['cabin']) & (new_df['name'] != dff.loc[0]['name'])]
for i in range(1, len(dff)):
    dff2 = pd.concat([dff2, new_df.loc[(new_df['cabin'] == dff.loc[i]['cabin']) & (new_df['name'] != dff.loc[i]['name'])]])
dff2
```
