## бе 1

Decision Tree Classifier

```python
# Имеем 2 класса - (1) - класс узла, (2) - класс дерева классификации
# (1): Описание узлов нашего дерева. Имеем только конструктор.
# (индекс характеристик, порог, право\лево(для доступа в правому\левому дочерним элементам),
# прирост информации(хранит инфу, полученную в рез-те разделения, обозначенную конкретным узлом) и значение).

class Node():
  def __init__(self, feature_index=None, threshold=None, right=None, left=None,  info_gain=None, value=None):
    self.feature_index = feature_index
    self.threshold = threshold
    self.right = right
    self.left = left
    self.info_gain = info_gain
    self.value = value

# (2)
# Метод построения дерева, метод рассчета энтропии, метод прогнозирования, метод расщепления и тд.

# Констуркор.
# root - корень (нужен для обхода дерева). 
# Оставшиеся два параметра - условия остановки.
# Минимальное разделение выборки и максимальная глубина. Есди выполняется критерий остановки -
# мы больеш не разбиваем узел, мы считаем его листовым.


class DecisionTreeClassifier():
  def __init__(self, min_samples_split=2, max_depth=2):
    self.root = None
    self.min_samples_split = min_samples_split
    self.max_depth = max_depth

# Далее самая важная функция в этом классе. Функция построения дерева. 
# Это рекурсивная фукнция для построения двоичного дерева.

  def build_tree(self, dataset, curr_depth=0):

    X, Y = dataset[:,:-1], dataset[:,-1]
    num_samples, num_features = np.shape(X)

    # Проверка условий остановки, далее мы проверяем, больше ли прирост информации, соотв. этому узлу, чем 0
    # (это делается потому что если ноль - узел чистый).
    # Далее создаем левое поддерево и правое поддерево. После того как поддеревья сделаны, нам нужно вернуть
    # узел, и этот узел - узел решения, а значит мы должны передать пороговое значение индекса функции threshold
    # Далее - находим конечный узел. Функция вычисления знаяения.

    if num_samples>=self.min_samples_split and curr_depth<=self.max_depth:
      best_split = self.get_best_split(dataset, num_samples, num_features)
      if best_split["info_gain"]>0:
        left_subtree = self.build_tree(best_split["dataset_left"], curr_depth+1)
        right_subtree = self.build_tree(best_split["dataset_right"], curr_depth+1)
        return Node(best_split["feature_index"], best_split["threshold"], left_subtree, right_subtree, best_split["info_gain"])
        
    leaf_value = self.calculate_leaf_value(Y)
    return Node(value=leaf_value)

  # Поиск наилучшей функции разделения.      
  # Возвращает словарь.

  def get_best_split(self, dataset, num_samples, num_features):
    best_split = {}
    max_info_gain = -float("inf") # Инициализировали как -бесконечность, т.к. надо максимизировать прирост инфы.
    
    # Необходимо перебрать все функции, и внутри этого цикла мы должны пройти через все возможные поддерево
    # значения. Теперь функции - действ. знаечния и существует бесконечное количество действ. чисел между
    # между номерами, поэтому нет смысла перебирать все возможные действительные числа.
    # Вместо этого просто проходим через все возможные значения. 
    
    # np.uniqie просто возвращает уникальные значения.

    for feature_index in range(num_features):
      feature_values = dataset[:, feature_index]
      possible_thresholds = np.unique(feature_values)

            # Во втором цикле разделим набор данных на основе текущего индекса и текущего порога.
            # На этом этапе у нас есть левый набор данных => запишем набор данных.
            # Убедимся, что они не пусты. Далее просто извлекаем целевые значения функции (y).
            # после нужно вычислить прирост информации. Проверяем, при необходимости обновляем разделение (split)
            # Dернем best_split.

      for threshold in possible_thresholds:
        dataset_left, dataset_right = self.split(dataset, feature_index, threshold)
        if len(dataset_left)>0 and len(dataset_right)>0:

          y, left_y, right_y = dataset[:, -1], dataset_left[:, -1], dataset_right[:, -1]
          curr_info_gain = self.information_gain(y, left_y, right_y, 'gini')

          if curr_info_gain>max_info_gain:
            best_split['feature_index'] = feature_index
            best_split['threshold'] = threshold
            best_split['dataset_left'] = dataset_left
            best_split['dataset_right'] = dataset_right
            best_split['info_gain'] = curr_info_gain
            max_info_gain = curr_info_gain

    return best_split


  # Функция разделения. Она берет набор данных, индекс, пороговое значение и разбивает его на  части
  # 1 часть будет идти к левому дочернему элементу, 2 часть - к правому.

  def split(self, dataset, feature_index, threshold):

    dataset_left = np.array([row for row in dataset if row[feature_index] <= threshold])
    dataset_right = np.array([row for row in dataset if row[feature_index] > threshold])
      
    return dataset_left, dataset_right

  # Функция вычисления прироста информации. Просто вычисляет объединенную инормацию дочерних из родительского.
  # Используем веса - относительыне размеры дочерних узлов по отноению к родительским.

  def information_gain(self, parent, l_child, r_child, mode='entropy'):

        # Энтропия и gini(делает в принципе то же самое, но формула немного отличается, мы удалили логарифм).

    weight_l = len(l_child) / len(parent)
    weight_r = len(r_child) / len(parent)

    if mode=='gini':
      gain = self.gini_index(parent) - (weight_l*self.gini_index(l_child) + weight_r*self.gini_index(r_child))
    else:
      gain = self.entropy(parent) - (weight_l*self.entropy(l_child) + weight_r*self.entropy(r_child))
    return gain

    # Следующие две функции - вычисление энтропии и gini.

  def entropy(self, y):
        
    class_labels = np.unique(y)
    entropy = 0
    for cls in class_labels:
      p_cls = len(y[y == cls]) / len(y)
      entropy += -p_cls * np.log2(p_cls)
    return entropy
    
  def gini_index(self, y):
        
    class_labels = np.unique(y)
    gini = 0
    for cls in class_labels:
      p_cls = len(y[y == cls]) / len(y)
      gini += p_cls**2
    return 1 - gini

  # Далее - функция вычисления конечного узла. Это в принципе просто класс большинства, присутствующий
  # в этом конкретном узле (найдем наиболее часто встречающийся элемент).

  def calculate_leaf_value(self, Y):
        
    Y = list(Y)
    return max(Y, key=Y.count)

  # Следующая функция - визуализация дерева.

  def print_tree(self, tree=None, indent=' '):
        
    if not tree:
      tree = self.root

    if tree.value is not None:
      print(tree.value)

    else:
      print("X_"+str(tree.feature_index), '<=', tree.threshold, '?', tree.info_gain)
      print("%sleft:" % (indent), end='')
      self.print_tree(tree.left, indent + indent)
      print("%sright:" % (indent), end='')
      self.print_tree(tree.right, indent + indent)

  # Функция обучения.
  # Сначала объединим X, Y, потом вызываем функцию построения дерева.

  def fit(self, X, Y):
    dataset = np.concatenate((X, Y), axis=1)
    self.root = self.build_tree(dataset)

  # Функция прогнозирвоания.
  def predict(self, X):
        
    preditions = [self.make_prediction(x, self.root) for x in X]
    return preditions

  # Изначально просто передаем корневой узел, если значение не none(если узел конечный) => возврат значения


  def make_prediction(self, x, tree):
        
    if tree.value!=None: return tree.value
    
    feature_val = x[tree.feature_index]

    if feature_val<=tree.threshold:
      return self.make_prediction(x, tree.left)

    else:
      return self.make_prediction(x, tree.right)
```

```python
# Определние индекса, порог, право-лево {уменьшение дисперсии слева и справа}, значение

class Node():
  def __init__(self, feature_index=None, threshold=None, left=None, right=None, var_red=None, value=None):

    self.feature_index = feature_index
    self.threshold = threshold
    self.left = left
    self.right = right
    self.var_red = var_red
    self.value = value

class DecisionTreeRegressor():
  # Минимальное разделение выборки, мкксимальная глубина - критерий останова.
  # Нам важно иметь правильную начальную точку(отправную).
  def __init__(self, min_samples_split=2, max_depth=2):
    self.root = None
    self.min_samples_split = min_samples_split
    self.max_depth = max_depth

  # Метод построения дерева - самая важная функция (также рекурсивная).       
  def build_tree(self, dataset, curr_depth=0):  
    X, Y = dataset[:,:-1], dataset[:,-1] # Отделили признаки от целевой переменной
    num_samples, num_features = np.shape(X)
    best_split = {}
  # Проверяем, соблюдаются ли критерии остановки. Если нет - ищем наилучшее разделение для текущего набора данных.
  # Если уменьшение дисперсии за счет лучшего разделения больше 0, мы разделяем набор данных.
  # После чего рекурсивно вызываем функцию build_tree для левого и правого поддеревьев.
  # Нам нужно передать пороговое значение индекса и уменьшение дисперсии, соотв. конкретному узлу, как только мы 
  # достигли самого нижнего уровня.
    if num_samples>=self.min_samples_split and curr_depth<=self.max_depth:
      best_split = self.get_best_split(dataset, num_samples, num_features)
      if best_split["var_red"]>0:
        left_subtree = self.build_tree(best_split["dataset_left"], curr_depth+1)

        right_subtree = self.build_tree(best_split["dataset_right"], curr_depth+1)

        return Node(best_split["feature_index"], best_split["threshold"], left_subtree, right_subtree, best_split["var_red"])
    
    # Далее необходимо вычислить значение выхода, для этого - calculate_leaf_value.
    # После того как это сделано, нам надо просто создать конечный узел.   
    
    leaf_value = self.calculate_leaf_value(Y)
    return Node(value=leaf_value)
    
  # Функция наилучшего разделения возвращает словарь. 


  def get_best_split(self, dataset, num_samples, num_features):

    best_split = {} # вот он собственно этот словарь (пока пустой) 
    max_var_red = -float("inf")

    # Цикл через все функции, внутри этого цикла мы должны пройти через все возможные пороговые значения.
    # Проходим через только существующие значения, которыое есть в наборе данных и тд и тп
    # np.unique - уникальные значения
  
    for feature_index in range(num_features):
      feature_values = dataset[:, feature_index]
      possible_thresholds = np.unique(feature_values)

      # Прежде всего разделяем набор данных на основе текущего индекса (и порога),  
      # подобные объяснения выше. 
      # (Имеем левый набор данных, убеждаемся, что этот набор не пуст. После подтвеждения просто извлекаем целевые значения)
      # После этого нам нужно вычислить прирост информации (дисперсия) --- variance_reduction.

      for threshold in possible_thresholds:
        dataset_left, dataset_right = self.split(dataset, feature_index, threshold)

        if len(dataset_left)>0 and len(dataset_right)>0:
          y, left_y, right_y = dataset[:, -1], dataset_left[:, -1], dataset_right[:, -1]

          # Как только получили прирост инфы, нужно проверить и при необходимости обновить наше наилучшее разделение.


          curr_var_red = self.variance_reduction(y, left_y, right_y)
          if curr_var_red>max_var_red:
            best_split["feature_index"] = feature_index
            best_split["threshold"] = threshold
            best_split["dataset_left"] = dataset_left
            best_split["dataset_right"] = dataset_right
            best_split["var_red"] = curr_var_red
            max_var_red = curr_var_red

    # Вернем лучшее разделение.

    return best_split
    
  def split(self, dataset, feature_index, threshold):

    dataset_left = np.array([row for row in dataset if row[feature_index]<=threshold])
    dataset_right = np.array([row for row in dataset if row[feature_index]>threshold])
    return dataset_left, dataset_right
    
  def variance_reduction(self, parent, l_child, r_child):

    # Просто вычитаем объединенную дисперсию левого и правого дочерних деревьев из дисперсии родительского.
    # Необходимо использовать веса, потому что количество элементов(точек, объектов) в родительском
    # больше чем в каждом из дочерних. Поэтому нам нужно масштабировать (коэф - относительный размер дочерних по отн к родителю)


    weight_l = len(l_child) / len(parent)
    weight_r = len(r_child) / len(parent)
    reduction = np.var(parent) - (weight_l * np.var(l_child) + weight_r * np.var(r_child))
    return reduction
    

  # Вычисление среднего значения всех точек данных в конкретном узле.  

  def calculate_leaf_value(self, Y):

    val = np.mean(Y)
    return val

  # Визуализация дерева -просто через обход дерева.

  def print_tree(self, tree=None, indent=" "):

    if not tree:
      tree = self.root

    if tree.value is not None:
      print(tree.value)

    else:
      print("X_"+str(tree.feature_index), "<=", tree.threshold, "?", tree.var_red)
      print("%sleft:" % (indent), end="")
      self.print_tree(tree.left, indent + indent)
      print("%sright:" % (indent), end="")
      self.print_tree(tree.right, indent + indent)
    
  # Обучение.   
  # Сначала объединим X, Y, потом вызываем функцию построения дерева.

  def fit(self, X, Y):

    dataset = np.concatenate((X, Y), axis=1)
    self.root = self.build_tree(dataset)
        
  def make_prediction(self, x, tree):

    # Эта функция в основном берет одну точку данных находит значение, соотв. этой точке
    # (если значение не none(если узел конечный) => возврат значения)
    # Выполняем рекурсивно, пока не достигнем конечного узгла - затем просто возвращаем значение в этом конечном узле

    if tree.value!=None: return tree.value
    feature_val = x[tree.feature_index]
    if feature_val<=tree.threshold:
      return self.make_prediction(x, tree.left)
    else:
      return self.make_prediction(x, tree.right)


  # Функция прогнозирования.
  # Может принимать целый массив точек данных и она будет передавать каждую точку make_prediction и выводить массив целевых значений

  def predict(self, X):

    preditions = [self.make_prediction(x, self.root) for x in X]
    return preditions
```
